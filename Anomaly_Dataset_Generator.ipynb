```python
import pandas as pd
import os
```

```python
# データが保存されているパスと保存先のパス
base_data_path = #波形データが入っているパスを指定
output_folder = #アウトプット用のパスを指定

# 解析データのパス
analysis_data_path = #解析データのパスを指定

# 解析データの一覧を取得
analysis_files = os.listdir(analysis_data_path)
analysis_files = [file for file in analysis_files if file.endswith('.xlsx')]

# 各解析データについてループ処理
for analysis_file in analysis_files:
    # 解析データを読み込む
    analysis_data = pd.read_excel(os.path.join(analysis_data_path, analysis_file), header=None)
    
    # 各行の処理
    for index, row in analysis_data.iterrows():
        # 管番号の生成
        col1 = str(row[0]).zfill(3)  # 1列目の数値
        col2 = str(row[1]).zfill(3)  # 2列目の数値
        tube_number = f"{col1}-{col2}"
        
        col4 = row[3]
        col17 = row[16]
        col12 = row[11]
        
        if col4 == 0:
            continue
        
        if col4 != 0 and col17 != 0:
            continue
        
        # 欠陥位置の計算
        defect_position = col12 / 2
        
        # 解析データに対応するフォルダとファイル名を作成
        folder_name = analysis_file.split('.')[0]
        data_folder_path = os.path.join(base_data_path, folder_name)
        
        # 欠陥位置を中心に-100～+100の範囲でデータを切り取り
        try:
            data = pd.read_excel(os.path.join(data_folder_path, f"{tube_number}.xlsx"), header=None)
            start_index = max(0, int(defect_position) - 99)
            end_index = min(len(data), int(defect_position) + 100)
            cut_data = data.iloc[start_index:end_index]
            
            # 出力ファイル名
            output_filename = f"{os.path.splitext(analysis_file)[0]}_{tube_number}_{int(defect_position)}.xlsx"
            output_path = os.path.join(output_folder, output_filename)
            
            # データを保存
            cut_data.to_excel(output_path, index=False, header=False)
            print(f"Saved {output_filename}")
            
        except Exception as e:
            print(f"Error processing {tube_number}: {str(e)}")
```
