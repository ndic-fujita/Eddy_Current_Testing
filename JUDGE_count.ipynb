```python
import os
import pandas as pd
from tkinter import Tk, filedialog
```

```python
# 関数定義：条件ごとのカウント
def count_conditions(data):
    counts = {
        "条件①": 0,
        "条件②": 0,
        "条件③": 0,
        "条件④": 0,
        "条件⑤": 0,
        "条件⑥": 0,
        "条件⑦": 0
    }
    
    num_columns = len(data.columns)  # 列数を取得
    
    for row_index in range(len(data)):
        if num_columns > 3:
            col4_value = data.iloc[row_index, 3]  # 4番目の列の値
        else:
            col4_value = None
        
        if num_columns > 16:
            col17_value = data.iloc[row_index, 16]  # 17番目の列の値
        else:
            col17_value = None
        
        if num_columns > 23:
            col24_value = data.iloc[row_index, 23]  # 24番目の列の値
        else:
            col24_value = None
        
        if col4_value is not None:
            if col4_value == 0:
                counts["条件①"] += 1
            elif col4_value == 1 and col17_value is not None and col17_value == 0:
                counts["条件②"] += 1
            elif col4_value == 2:
                if col17_value is not None and col17_value == 0:
                    counts["条件③"] += 1
                if col24_value is not None and col24_value == 0:
                    counts["条件④"] += 1
                if col17_value is not None and col17_value != 0:
                    counts["条件⑥"] += 1
                if col24_value is not None and col24_value != 0:
                    counts["条件⑦"] += 1
            elif col4_value >= 3:
                counts["条件⑤"] += 1
        
    return counts

# メイン処理
def main(folder_path):
    total_counts = {
        "条件①": 0, #①4番目の列の値が「0」
        "条件②": 0, #②4番目の列の値が「1」且つ17番目の列の値が「0」
        "条件③": 0, #③4番目の列の値が「2」且つ17番目の列の値が「0」
        "条件④": 0, #④4番目の列の値が「2」且つ24番目の列の値が「0」
        "条件⑤": 0, #⑤4番目の列の値が「3」以上
        "条件⑥": 0, #⑥4番目の列の値が「2」且つ17番目の列の値が「0」以外
        "条件⑦": 0  #⑦4番目の列の値が「2」且つ24番目の列の値が「0」以外
    }

    # 指定したフォルダ内のすべてのExcelファイルに対して処理を行う
    for filename in os.listdir(folder_path):
        if filename.endswith(".xlsx") or filename.endswith(".xls"):
            file_path = os.path.join(folder_path, filename)
            print(f"ファイル: {filename}")
            
            # Excelファイルを読み込む
            try:
                df = pd.read_excel(file_path, header=None)  # ヘッダーがない場合はNoneにする
                counts = count_conditions(df)
                
                # 各ファイルのカウントをトータルに加算する
                for condition, count in counts.items():
                    total_counts[condition] += count
                
                # ファイルごとの結果を出力する
                for condition, count in counts.items():
                    print(f"{condition}　{count}個")
                print("-------------------------")
                
            except Exception as e:
                print(f"エラー: {e}")
    
    # トータルの結果を出力する
    print("トータル結果:")
    for condition, count in total_counts.items():
        print(f"{condition}　{count}個")

# フォルダのパスを指定して実行する例
if __name__ == "__main__":
    folder_path = r"C:\Users\r-fujita\Desktop\教師データJUDGE"
    main(folder_path)
```
