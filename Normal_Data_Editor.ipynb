```python
import pandas as pd
import os
import random
```

```python
# 
base_data_path = r"C:\Users\r-fujita\Desktop\Phase2\Excel"
output_folder = r"C:\Users\r-fujita\Desktop\Phase2\dataset_normal"

# 
analysis_data_path = r"C:\Users\r-fujita\Desktop\Phase2\JUDGE"

# 
analysis_files = os.listdir(analysis_data_path)
analysis_files = [file for file in analysis_files if file.endswith('.xlsx')]

# 
def select_random_points(center, width, data_length):
    start = max(0, center - width // 2)
    end = min(data_length, start + width)
    return list(range(start, end))

# 
for analysis_file in analysis_files:
    # 
    analysis_data = pd.read_excel(os.path.join(analysis_data_path, analysis_file), header=None)
    
    # 
    for index, row in analysis_data.iterrows():
        # 
        col1 = str(row[0]).zfill(3)  # 1
        col2 = str(row[1]).zfill(3)  # 2
        tube_number = f"{col1}-{col2}"
        
        col4 = row[3]
        
        # 140
        if col4 == 0:
            # 
            folder_name = analysis_file.split('.')[0]
            data_folder_path = os.path.join(base_data_path, folder_name)
            data_file_path = os.path.join(data_folder_path, f"{tube_number}.xlsx")
            
            # 
            try:
                data = pd.read_excel(data_file_path)
                data_length = len(data)
                
                if data_length <= 200:
                    continue  # 200
                
                # 5100
                for _ in range(5):
                    center_position = random.randint(50, data_length - 50)
                    selected_rows = select_random_points(center_position, 200, data_length)
                    
                    # 
                    output_filename = f"{os.path.splitext(analysis_file)[0]}_{tube_number}_{center_position}.xlsx"
                    output_path = os.path.join(output_folder, output_filename)
                    
                    # Excel
                    cut_data = data.iloc[selected_rows]
                    cut_data.to_excel(output_path, index=False, header=False)
                    print(f"Saved {output_filename}")
            
            except Exception as e:
                print(f"Error processing {tube_number}: {str(e)}")
```
