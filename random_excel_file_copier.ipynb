```python
import os
import shutil
import random

def copy_random_excel_files(src_dir, dst_dir, num_files):
    # 
    if not os.path.exists(dst_dir):
        os.makedirs(dst_dir)
    
    # src_dir
    excel_files = [f for f in os.listdir(src_dir) if f.endswith('.xlsx')]
    
    # num_files
    if len(excel_files) < num_files:
        raise ValueError(f"Source directory does not contain enough Excel files. Found {len(excel_files)}, but need {num_files}.")
    
    # 
    random_files = random.sample(excel_files, num_files)
    
    # 
    for file in random_files:
        src_file = os.path.join(src_dir, file)
        dst_file = os.path.join(dst_dir, file)
        shutil.copy(src_file, dst_file)
        print(f"Copied {file} to {dst_dir}")

# 
source_directory = r"C:\Users\r-fujita\Desktop\Phase2\dataset_normal()"
destination_directory = r"C:\Users\r-fujita\Desktop\Phase2\dataset_normal"
number_of_files_to_copy = 15000

copy_random_excel_files(source_directory, destination_directory, number_of_files_to_copy)
```
