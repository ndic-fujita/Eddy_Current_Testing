```python
import os
import shutil
import random

# デバイスフォルダのパス
device_folder_path = r"C:\Users\r-fujita\Desktop\ETデータ整理用"
# NG_test フォルダのパス
ng_test_path = r"C:\Users\r-fujita\Desktop\ET_Dataset\NG_test"
# OK_test フォルダのパス
ok_test_path = r"C:\Users\r-fujita\Desktop\ET_Dataset\OK_test"
# train フォルダのパス
train_path = r"C:\Users\r-fujita\Desktop\ET_Dataset\train"

# デバイスフォルダ内のフォルダをリスト化
device_folders = os.listdir(device_folder_path)

# デバイスフォルダごとに処理を行う
for device_folder in device_folders:
    # タイプフォルダのパス
    type_folder_path = os.path.join(device_folder_path, device_folder)
    # タイプフォルダ内のフォルダをリスト化
    type_folders = os.listdir(type_folder_path)

    # NG_test フォルダ内にデバイスフォルダと同じ名前のフォルダを作成
    ng_device_folder_path = os.path.join(ng_test_path, device_folder)
    os.makedirs(ng_device_folder_path, exist_ok=True)

    # OK_test と train フォルダ内にデバイスフォルダと同じ名前のフォルダを作成
    ok_train_device_folder_path = os.path.join(ok_test_path, device_folder)
    os.makedirs(ok_train_device_folder_path, exist_ok=True)
    train_device_folder_path = os.path.join(train_path, device_folder)
    os.makedirs(train_device_folder_path, exist_ok=True)

    # タイプフォルダごとに処理を行う
    for type_folder in type_folders:
        # タイプフォルダのパス
        type_folder_path = os.path.join(device_folder_path, device_folder, type_folder)
        
        # '内面'の場合
        if type_folder == '内面':
            # '内面'フォルダのパス
            ng_type_folder_path = os.path.join(ng_device_folder_path, '内面')
            # フォルダを作成するかどうかを確認
            if not os.path.exists(ng_type_folder_path):
                print("内面フォルダを作成します:", ng_type_folder_path)  # デバッグステートメント
                os.makedirs(ng_type_folder_path)
            else:
                print("内面フォルダはすでに存在します:", ng_type_folder_path)  # デバッグステートメント
            # タイプフォルダの中身を'内面'フォルダにコピー
            for item in os.listdir(type_folder_path):
                print("Copying", item, "to", ng_type_folder_path)  # デバッグステートメント
                shutil.copy(os.path.join(type_folder_path, item), ng_type_folder_path)

        
       # '外面'の場合
        elif type_folder == '外面':
            # '外面'フォルダのパス
            ng_type_folder_path = os.path.join(ng_device_folder_path, '外面')
            # フォルダを作成するかどうかを確認
            if not os.path.exists(ng_type_folder_path):
                print("外面フォルダを作成します:", ng_type_folder_path)  # デバッグステートメント
                os.makedirs(ng_type_folder_path)
            else:
                print("外面フォルダはすでに存在します:", ng_type_folder_path)  # デバッグステートメント
            # タイプフォルダの中身を'外面'フォルダにコピー
            for item in os.listdir(type_folder_path):
                print("Copying", item, "to", ng_type_folder_path)  # デバッグステートメント
                shutil.copy(os.path.join(type_folder_path, item), ng_type_folder_path)

        # '全面'の場合
        elif type_folder == '全面':
            # '全面'フォルダのパス
            ng_type_folder_path = os.path.join(ng_device_folder_path, '全面')
            # フォルダを作成するかどうかを確認
            if not os.path.exists(ng_type_folder_path):
                print("全面フォルダを作成します:", ng_type_folder_path)  # デバッグステートメント
                os.makedirs(ng_type_folder_path)
            else:
                print("全面フォルダはすでに存在します:", ng_type_folder_path)  # デバッグステートメント
            # タイプフォルダの中身を'全面'フォルダにコピー
            for item in os.listdir(type_folder_path):
                print("Copying", item, "to", ng_type_folder_path)  # デバッグステートメント
                shutil.copy(os.path.join(type_folder_path, item), ng_type_folder_path)

        # '正常'の場合、'異常'の場合、'打痕'の場合
        elif type_folder in ['正常', '異常', '打痕']:
            # 対応するフォルダのパス
            folder_name = {'正常': 'OK_test', '異常': 'OK_test', '打痕': 'OK_test'}.get(type_folder)
            # フォルダを作成するかどうかを確認
            target_path = os.path.join(ok_test_path if folder_name == 'OK_test' else train_path, device_folder, type_folder)
            if not os.path.exists(target_path):
                print(f"{type_folder}フォルダを作成します:", target_path)  # デバッグステートメント
                os.makedirs(target_path)
            else:
                print(f"{type_folder}フォルダはすでに存在します:", target_path)  # デバッグステートメント

            # trainフォルダ内にデバイスフォルダを作成し、その中に'正常', '異常', '打痕'フォルダを作成
            train_device_folder_path = os.path.join(train_path, device_folder)
            if not os.path.exists(train_device_folder_path):
                print(f"{device_folder}フォルダを作成します:", train_device_folder_path)  # デバッグステートメント
                os.makedirs(train_device_folder_path)
            else:
                print(f"{device_folder}フォルダはすでに存在します:", train_device_folder_path)  # デバッグステートメント
            train_device_type_folder_path = os.path.join(train_device_folder_path, type_folder)
            if not os.path.exists(train_device_type_folder_path):
                print(f"{type_folder}フォルダを作成します:", train_device_type_folder_path)  # デバッグステートメント
                os.makedirs(train_device_type_folder_path)
            else:
                print(f"{type_folder}フォルダはすでに存在します:", train_device_type_folder_path)  # デバッグステートメント

            # タイプフォルダの中身を振り分け
            files = os.listdir(type_folder_path)
            random.shuffle(files)
            split_index = int(0.2 * len(files)) if folder_name == 'OK_test' else len(files)
            target_files = files[:split_index] if folder_name == 'OK_test' else files[split_index:]

            # ファイルを振り分け
            for item in target_files:
                print("Copying", item, "to", target_path)  # デバッグステートメント
                shutil.copy(os.path.join(type_folder_path, item), target_path)
            for item in files:
                if item not in target_files:
                    print("Copying", item, "to", train_device_type_folder_path)  # デバッグステートメント
                    shutil.copy(os.path.join(type_folder_path, item), train_device_type_folder_path)
```
