```python
import os
import pandas as pd
from shutil import copy

def convert_excel_to_csv(source_folder, destination_folder):
    # ディレクトリが存在しない場合は作成する
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # ソースフォルダ内のすべてのファイルに対して処理を行う
    for filename in os.listdir(source_folder):
        if filename.endswith('.xlsx') or filename.endswith('.xls'):
            # Excelファイルのパス
            excel_file_path = os.path.join(source_folder, filename)
            
            # Excelファイルを読み込む
            excel_data = pd.read_excel(excel_file_path)
            
            # CSVファイルのパスを生成（拡張子を変更）
            csv_file_path = os.path.join(destination_folder, filename.rsplit('.', 1)[0] + '.csv')
            
            # CSVファイルとして保存
            excel_data.to_csv(csv_file_path, index=False)
            
            # 移動元のExcelファイルを移動先にコピーする
            copy(excel_file_path, destination_folder)
            
            print(f'{filename} を CSV に変換してコピーしました。')

    print('すべてのファイルの変換とコピーが完了しました。')

# 使用例
if __name__ == "__main__":
    source_folder = '/path/to/source/folder'  # 移動元のフォルダパス
    destination_folder = '/path/to/destination/folder'  # 移動先のフォルダパス

    convert_excel_to_csv(source_folder, destination_folder)


```
